```python
#!pip install gdown
# Importing necessary libraries
#import pandas as pd
#from sklearn.model_selection import train_test_split
#from sklearn.preprocessing import LabelEncoder, StandardScaler
#from sklearn.tree import DecisionTreeClassifier
#from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
#from sklearn.linear_model import LogisticRegression
#from sklearn.svm import SVC
#from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report
#from sklearn.model_selection import cross_val_score
#import seaborn as sns
#import matplotlib.pyplot as plt
#from sklearn.metrics import precision_recall_curve

# Load the data
#from google.colab import files
#uploaded = files.upload()
#data = pd.read_csv("shopping_trends.csv")
#from google.colab import drive
#https://drive.google.com/file/d/18I51EUYsNq4gvk2t7vpIGXvfIu8_4a0p/view?usp=drive_link
# Install gdown to download the file from Google Drive
!pip install gdown
!pip install openpyxl
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report
from sklearn.model_selection import cross_val_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Import gdown
import gdown
import openpyxl
# Replace with your File ID
file_id = '18I51EUYsNq4gvk2t7vpIGXvfIu8_4a0p'
url = f'https://drive.google.com/uc?id={file_id}'

# Download the file to your Colab environment
gdown.download(url, 'shopping_trends.csv', quiet=False)

# Load the dataset
#df = pd.read_excel('dataset.xlsx')
df = pd.read_csv('shopping_trends.csv')

# Display the first few rows to ensure it's loaded correctly
df.head()



# Data Preprocessing
# Assuming 'Subscription Status' is the target variable
le = LabelEncoder()
df['Subscription Status'] = le.fit_transform(df['Subscription Status'])

# One-hot encode categorical columns
data = pd.get_dummies(df, columns=['Discount Applied', 'Promo Code Used'])

# Selecting relevant features
features = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases',
            'Discount Applied_Yes', 'Discount Applied_No', 'Promo Code Used_Yes', 'Promo Code Used_No']

# Extracting the features to be normalized
features_to_normalize = data[features]

# Normalize the features using StandardScaler
scaler = StandardScaler()
normalized_features = scaler.fit_transform(features_to_normalize)

# Replace the original features with the normalized ones
data[features] = normalized_features

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[features], data['Subscription Status'], test_size=0.2, random_state=42)

# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
lr_accuracy = accuracy_score(y_test, lr_predictions)

# Support Vector Machines (SVM)
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_predictions)

# Gradient Boosting
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
gb_predictions = gb_model.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_predictions)

# Additional Evaluation Metrics
# Confusion Matrix
cm_dt = confusion_matrix(y_test, dt_predictions)
cm_rf = confusion_matrix(y_test, rf_predictions)
cm_lr = confusion_matrix(y_test, lr_predictions)
cm_svm = confusion_matrix(y_test, svm_predictions)
cm_gb = confusion_matrix(y_test, gb_predictions)

print("\nConfusion Matrix for Decision Tree:")
print(cm_dt)
print("\nConfusion Matrix for Random Forest:")
print(cm_rf)
print("\nConfusion Matrix for Logistic Regression:")
print(cm_lr)
print("\nConfusion Matrix for SVM:")
print(cm_svm)
print("\nConfusion Matrix for Gradient Boosting:")
print(cm_gb)

# Precision, Recall, F1-Score
precision_dt = precision_score(y_test, dt_predictions)
recall_dt = recall_score(y_test, dt_predictions)
f1_dt = f1_score(y_test, dt_predictions)

precision_rf = precision_score(y_test, rf_predictions)
recall_rf = recall_score(y_test, rf_predictions)
f1_rf = f1_score(y_test, rf_predictions)

precision_lr = precision_score(y_test, lr_predictions)
recall_lr = recall_score(y_test, lr_predictions)
f1_lr = f1_score(y_test, lr_predictions)

precision_svm = precision_score(y_test, svm_predictions)
recall_svm = recall_score(y_test, svm_predictions)
f1_svm = f1_score(y_test, svm_predictions)

precision_gb = precision_score(y_test, gb_predictions)
recall_gb = recall_score(y_test, gb_predictions)
f1_gb = f1_score(y_test, gb_predictions)

# ROC-AUC Curve
roc_auc_dt = roc_auc_score(y_test, dt_predictions)
fpr_dt, tpr_dt, _ = roc_curve(y_test, dt_predictions)

roc_auc_rf = roc_auc_score(y_test, rf_predictions)
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_predictions)

roc_auc_lr = roc_auc_score(y_test, lr_predictions)
fpr_lr, tpr_lr, _ = roc_curve(y_test, lr_predictions)

roc_auc_svm = roc_auc_score(y_test, svm_predictions)
fpr_svm, tpr_svm, _ = roc_curve(y_test, svm_predictions)

roc_auc_gb = roc_auc_score(y_test, gb_predictions)
fpr_gb, tpr_gb, _ = roc_curve(y_test, gb_predictions)

# Classification Report
classification_report_dt = classification_report(y_test, dt_predictions)
classification_report_rf = classification_report(y_test, rf_predictions)
classification_report_lr = classification_report(y_test, lr_predictions)
classification_report_svm = classification_report(y_test, svm_predictions)
classification_report_gb = classification_report(y_test, gb_predictions)

# Cross-Validation
cross_val_scores_dt = cross_val_score(dt_model, data[features], data['Subscription Status'], cv=5)
cross_val_scores_rf = cross_val_score(rf_model, data[features], data['Subscription Status'], cv=5)
cross_val_scores_lr = cross_val_score(lr_model, data[features], data['Subscription Status'], cv=5)
cross_val_scores_svm = cross_val_score(svm_model, data[features], data['Subscription Status'], cv=5)
cross_val_scores_gb = cross_val_score(gb_model, data[features], data['Subscription Status'], cv=5)

# Print the results
print(f"\nDecision Tree Accuracy: {dt_accuracy}")
print(f"Random Forest Accuracy: {rf_accuracy}")
print(f"Logistic Regression Accuracy: {lr_accuracy}")
print(f"SVM Accuracy: {svm_accuracy}")
print(f"Gradient Boosting Accuracy: {gb_accuracy}")

# Print additional metrics for Decision Tree
print("\nAdditional Evaluation Metrics for Decision Tree:")
print(f"Precision: {precision_dt}")
print(f"Recall: {recall_dt}")
print(f"F1-Score: {f1_dt}")
print(f"ROC-AUC Score: {roc_auc_dt}")
print("\nClassification Report for Decision Tree:")
print(classification_report_dt)
print("\nCross-Validation Scores for Decision Tree:")
print(cross_val_scores_dt)

# Print additional metrics for Random Forest
print("\nAdditional Evaluation Metrics for Random Forest:")
print(f"Precision: {precision_rf}")
print(f"Recall: {recall_rf}")
print(f"F1-Score: {f1_rf}")
print(f"ROC-AUC Score: {roc_auc_rf}")
print("\nClassification Report for Random Forest:")
print(classification_report_rf)
print("\nCross-Validation Scores for Random Forest:")
print(cross_val_scores_rf)

# Print additional metrics for Logistic Regression
print("\nAdditional Evaluation Metrics for Logistic Regression:")
print(f"Precision: {precision_lr}")
print(f"Recall: {recall_lr}")
print(f"F1-Score: {f1_lr}")
print(f"ROC-AUC Score: {roc_auc_lr}")
print("\nClassification Report for Logistic Regression:")
print(classification_report_lr)
print("\nCross-Validation Scores for Logistic Regression:")
print(cross_val_scores_lr)

# Print additional metrics for SVM
print("\nAdditional Evaluation Metrics for SVM:")
print(f"Precision: {precision_svm}")
print(f"Recall: {recall_svm}")
print(f"F1-Score: {f1_svm}")
print(f"ROC-AUC Score: {roc_auc_svm}")
print("\nClassification Report for SVM:")
print(classification_report_svm)
print("\nCross-Validation Scores for SVM:")
print(cross_val_scores_svm)

# Print additional metrics for Gradient Boosting
print("\nAdditional Evaluation Metrics for Gradient Boosting:")
print(f"Precision: {precision_gb}")
print(f"Recall: {recall_gb}")
print(f"F1-Score: {f1_gb}")
print(f"ROC-AUC Score: {roc_auc_gb}")
print("\nClassification Report for Gradient Boosting:")
print(classification_report_gb)
print("\nCross-Validation Scores for Gradient Boosting:")
print(cross_val_scores_gb)

# Confusion Matrix Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Decision Tree')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Confusion Matrix Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Confusion Matrix Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Confusion Matrix Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - SVM')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Confusion Matrix Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm_gb, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Gradient Boosting')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC = {roc_auc_dt:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Random')
plt.title('ROC Curve - Decision Tree')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# ROC Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Random')
plt.title('ROC Curve - Random Forest')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# ROC Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {roc_auc_lr:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Random')
plt.title('ROC Curve - Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# ROC Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr_svm, tpr_svm, label=f'SVM (AUC = {roc_auc_svm:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Random')
plt.title('ROC Curve - SVM')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# ROC Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr_gb, tpr_gb, label=f'Gradient Boosting (AUC = {roc_auc_gb:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray', label='Random')
plt.title('ROC Curve - Gradient Boosting')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

# Feature Importance Plot for Decision Tree
if hasattr(dt_model, 'feature_importances_'):
    feature_importance_dt = dt_model.feature_importances_
    features_dt = data[features].columns
    importance_df_dt = pd.DataFrame({'Feature': features_dt, 'Importance': feature_importance_dt})
    importance_df_dt = importance_df_dt.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df_dt)
    plt.title('Feature Importance - Decision Tree')
    plt.show()

# Feature Importance Plot for Random Forest
if hasattr(rf_model, 'feature_importances_'):
    feature_importance_rf = rf_model.feature_importances_
    features_rf = data[features].columns
    importance_df_rf = pd.DataFrame({'Feature': features_rf, 'Importance': feature_importance_rf})
    importance_df_rf = importance_df_rf.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df_rf)
    plt.title('Feature Importance - Random Forest')
    plt.show()

# Feature Importance Plot for Logistic Regression (Coefficients)
if hasattr(lr_model, 'coef_'):
    feature_importance_lr = lr_model.coef_[0]
    features_lr = data[features].columns
    importance_df_lr = pd.DataFrame({'Feature': features_lr, 'Importance': feature_importance_lr})
    importance_df_lr['Importance'] = importance_df_lr['Importance'].abs()  # Use absolute values for coefficients
    importance_df_lr = importance_df_lr.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df_lr)
    plt.title('Feature Importance - Logistic Regression (Coefficients)')
    plt.show()

# Feature Importance Plot for SVM (Linear Kernel)
if hasattr(svm_model, 'coef_'):
    feature_importance_svm = svm_model.coef_[0]
    features_svm = data[features].columns
    importance_df_svm = pd.DataFrame({'Feature': features_svm, 'Importance': feature_importance_svm})
    importance_df_svm['Importance'] = importance_df_svm['Importance'].abs()  # Use absolute values for coefficients
    importance_df_svm = importance_df_svm.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df_svm)
    plt.title('Feature Importance - SVM (Linear Kernel)')
    plt.show()


# Feature Importance Plot for Gradient Boosting
if hasattr(gb_model, 'feature_importances_'):
    feature_importance_gb = gb_model.feature_importances_
    features_gb = data[features].columns
    importance_df_gb = pd.DataFrame({'Feature': features_gb, 'Importance': feature_importance_gb})
    importance_df_gb = importance_df_gb.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=importance_df_gb)
    plt.title('Feature Importance - Gradient Boosting')
    plt.show()
```
